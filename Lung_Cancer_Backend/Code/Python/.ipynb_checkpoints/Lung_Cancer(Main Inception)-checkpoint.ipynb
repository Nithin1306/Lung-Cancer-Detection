```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from keras.layers import Dense,Conv2D,MaxPooling2D,Flatten,BatchNormalization,Dropout
from keras.layers import Lambda
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img, img_to_array
from tensorflow.keras.models import Sequential
from glob import glob
```

```python
IMAGE_SIZE = [256,256]

train_path = r"C:\Users\allan\Desktop\Lung Cancer Backend\Data\train"
test_path = r"C:\Users\allan\Desktop\Lung Cancer Backend\Data\test"
valid_path = r"C:\Users\allan\Desktop\Lung Cancer Backend\Data\valid"
```

```python
inception = InceptionV3(input_shape=IMAGE_SIZE+[3],weights='imagenet',include_top=False)
```

```python
for layers in inception.layers:
    layers.trainable = False
```

```python
# Used to check how many class folders are there
folders = glob(r'C:\Users\allan\Desktop\Lung cancer\Lung Cancer Backend\Data\train\*')
```

```python
len(folders)
```

```python
x = Flatten()(inception.output)
```

```python
prediction = Dense(4,activation = 'softmax')(x)
```

```python
model = Model(inputs = inception.input , outputs = prediction)
```

```python
model.summary()
```

```python
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(r'C:\Users\allan\Desktop\Lung cancer\Lung Cancer Backend\Data\train',#Path of the dataset
                                                target_size = (256,256),
                                                batch_size = 32,
                                                class_mode = 'categorical')
```

```python
testing_set = test_datagen.flow_from_directory(r'C:\Users\allan\Desktop\Lung cancer\Lung Cancer Backend\Data\test',#Path of the dataset
                                                target_size = (256,256),
                                                batch_size = 32,
                                                class_mode = 'categorical')
```

```python
r = model.fit(
    training_set,
    validation_data = testing_set,
    epochs = 20
    
)
```

```python
model.save(f"../Model/1")
```

```python
model.save('C:/Users/allan/Desktop/Lung_Cancer/Model/Model_0.9576.h5')
```

```python
h = r.history
h.keys()
plt.plot(h['loss'] , 'go--' , c = 'red')
plt.plot(h['accuracy'] , 'go--' , color = 'Green')

plt.title('Loss vs Accuracy')
```
