```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input,Dense,Flatten
from keras.layers import Lambda
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img, img_to_array
from tensorflow.keras.models import Sequential
from glob import glob
```

```python
IMAGE_SIZE = [256,256]

train_path = r'C:\Users\allan\Desktop\Lung_Cancer\Data\train'
test_path = r'C:\Users\allan\Desktop\Lung_Cancer\Data\test'
valid_path = r'C:\Users\allan\Desktop\Lung_Cancer\Data\test'
```

```python
inception = InceptionV3(input_shape=IMAGE_SIZE+[3],weights='imagenet',include_top=False)
```

```python
for layers in inception.layers:
    layers.trainable = False
```

```python
# Used to check how many class folders are there
folders = glob(r'C:\Users\allan\Desktop\Lung_Cancer\Data\*')
folders
```

```python
x = Flatten()(inception.output)
```

```python
prediction = Dense(len(folders),activation = 'softmax')(x)
```

```python
model = Model(inputs = inception.input , outputs = prediction)
```

```python
model.summary()
```

```python
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = 'accuracy'
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
valid_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(r'C:\Users\allan\Desktop\Lung_Cancer\Data\train',#Path of the dataset
                                                target_size = (256,256),
                                                batch_size = 32,
                                                class_mode = 'categorical')
```

```python
testing_set = test_datagen.flow_from_directory(r'C:\Users\allan\Desktop\Lung_Cancer\Data\test',#Path of the dataset
                                                target_size = (256,256),
                                                batch_size = 32,
                                                class_mode = 'categorical')
```

```python
valid_set = valid_datagen.flow_from_directory(r'C:\Users\allan\Desktop\Lung_Cancer\Data\valid',#Path of the dataset
                                                target_size = (256,256),
                                                batch_size = 32,
                                                class_mode = 'categorical')
```

```python
#Plotting the images which is preprocessed in the training_set,testing_set,valid set

training_img , training_label = training_set.next()

testing_img , testing_label = testing_set.next()

validation_img , validation_label = valid_set.next()
```

```python
#Function to plot the images for each dataset which is been preprocessed

def plot(in_array , label):
    
    
    for idx,img in enumerate(in_array):
        if(idx<5):
            plt.figure(figsize=(2,2))
            plt.imshow(img)
            plt.show()
```

```python
#Plotting the images for Training images 

plot(training_img,training_label)
```

```python
#Plotting images for Testing Images

plot(testing_img,testing_label)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(256, 256, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
r = model.fit(
    training_set,
    validation_data = testing_set,
    epochs = 20,
    steps_per_epoch = len(training_set),
    validation_steps = len(testing_set)
)
```

```python
model.save('C:/Users/allan/Desktop/Lung_Cancer/Model/model.h5')
```

```python
h = r.history
h.keys()
```

```python
#PLotting the loss and accurancy

plt.plot(h['loss'] , 'go--' , c = 'red')
plt.plot(h['accuracy'] , 'go--' , color = 'Green')

plt.title('Loss vs Accurancy')
```
