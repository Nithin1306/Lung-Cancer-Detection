```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Input,Dense,Flatten
from keras.layers import Lambda
from tensorflow.keras.models import Model
from tensorflow.keras.applications.inception_v3 import InceptionV3
from tensorflow.keras.applications.inception_v3 import preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator,load_img, img_to_array
from tensorflow.keras.models import Sequential
from glob import glob
```

```python
IMAGE_SIZE = [256,256]

train_path = r"C:\Users\allan\Desktop\Skin_Data\train"
test_path = r"C:\Users\allan\Desktop\Skin_Data\test"
data_path = r"C:\Users\allan\Desktop\Skin_Data\data"
```

```python
inception = InceptionV3(input_shape=IMAGE_SIZE+[3],weights='imagenet',include_top=False)
```

```python
for layers in inception.layers:
    layers.trainable = False
```

```python
# Used to check how many class folders are there
folders = glob(r"C:\Users\allan\Desktop\Skin_Data\train\*")
```

```python
folders
```

```python
x = Flatten()(inception.output)
```

```python
prediction = Dense(len(folders),activation = 'softmax')(x)
```

```python
model = Model(inputs = inception.input , outputs = prediction)
```

```python
model.summary()
```

```python
model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)
```

```python
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
training_set = train_datagen.flow_from_directory(r"C:\Users\allan\Desktop\Skin_Data\train",#Path of the dataset
                                                target_size = (256,256),
                                                batch_size = 16,
                                                class_mode = 'categorical')
```

```python
testing_set = test_datagen.flow_from_directory(r"C:\Users\allan\Desktop\Skin_Data\test",#Path of the dataset
                                                target_size = (256,256),
                                                batch_size = 16,
                                                class_mode = 'categorical')
```

```python
r = model.fit(
    training_set,
    validation_data = testing_set,
    epochs = 20,
    steps_per_epoch = len(training_set),
    validation_steps = len(testing_set)
    
)
```

```python
model.save('C:/Users/allan/Desktop/Lung_Cancer/Model/Skin.h5')
```

```python
h = r.history
h.keys()
plt.plot(h['loss'] , 'go--' , c = 'red')
plt.plot(h['accuracy'] , 'go--' , color = 'Green')

plt.title('Loss vs Accurancy')
```
